```python
#similar artists are present in sinilar lists
import sys
import json
import re
import collections
import os
import itertools as it
import gensim
import pickle

path='../../data'
backgroundFiles=pickle.load(open('./SplitsInformation/filenamesBackGroundData.pkl','rb'))

playlistTrackMapping={}
trackPlaylistMapping={}
trackArtistMapping={}
artistTrackMapping={}
playlistArtistMapping={}
artistPlaylistMapping={}


filenames = os.listdir(path)
for filename in sorted(filenames):
    if filename.startswith("mpd.slice.") and filename.endswith(".json") and filename not in backgroundFiles:
        fullpath = os.sep.join((path, filename))
        f = open(fullpath)
        js = f.read()
        f.close()
        mpd_slice = json.loads(js)
        for playlist in mpd_slice['playlists']:
            trackSet=[]
            artistSet=[]
            playlistId=str(playlist['pid'])
            for track in playlist['tracks']:
                trackid=track['track_uri'].replace('spotify:track:','')
                trackArtist=track['artist_uri']
                trackSet.append(trackid)
                artistSet.append(trackArtist)
                artistSet.append(trackArtist)
                trackArtistMapping[trackid]=trackArtist
                #artistTrackMapping[trackArtist]=trackid
                artistTrackMapping.setdefault(trackArtist,[]).append(trackid)
                trackPlaylistMapping.setdefault(trackid,[]).append(playlistId)
            playlistTrackMapping[playlistId]=trackSet
            playlistArtistMapping[playlistId]=artistSet
            for art in set(artistSet):
                artistPlaylistMapping.setdefault(art,[]).append(playlistId)
```

```python
#similar lists have similar items -> Artist -Item -List -Artist
newrandomWalksArtistItem=[]
import random
pidsDone=0
for pid in playlistTrackMapping:
    pidCount=0
    while pidCount<=50: 
        iterPid=pid
        count=0
        rw=[]
        while count <=50:
            track=random.choice(playlistTrackMapping[pid])
            rw.append(track)
            artist=trackArtistMapping[track]
            iterPid = random.choice(artistPlaylistMapping[artist])
            count=count+1
        newrandomWalksArtistItem.append(rw)
        pidCount+=1
    pidsDone+=1
    if pidsDone%50000 ==0:
        print "processing "+ str(pidsDone) +'.....'
```
